```python
from dsc.query_engine import Query_Processor as dscQP

def dscquery(dsc_output, pstring, target, condition, groups):
    db = os.path.join(dsc_output, os.path.basename(os.path.normpath(dsc_output)) + ".db")
    qp = dscQP(db, target, condition, groups)
    pipelines = ['+'.join(x) for x in get_pipelines(pstring, qp.data['.groups'])]
    res = {x: qp.output_tables[x] for x in pipelines}
    return res

    
def is_defined_group(string):
    string = string.strip()
    return string.startswith('(') and string.endswith(')')


def read_groups(string):
    string = string.strip()
    string = string.replace('(', '')
    string = string.replace(')', '')
    return [x.strip() for x in string.strip().split(',')]

    
def get_pipelines(pstring, qp_groups):
    pipeline_steps = [x.strip() for x in pstring.split("*")]
    mpipelines = [""]
    for step in pipeline_steps:
        if step in qp_groups.keys():
            mpipelines = [f"{x} * {f}" for f in qp_groups[step] for x in mpipelines]
        elif is_defined_group(step):
            mpipelines = [f"{x} * {f}" for f in read_groups(step) for x in mpipelines]
        else:
            mpipelines = [f"{x} * {step}"  if x else f"{step}" for x in mpipelines]

    return [[y.strip() for y in x.split('*')] for x in mpipelines]
```
